```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib


matplotlib.rcParams['font.family'] = ['Times New Roman', 'SimSun']
matplotlib.rcParams['font.size'] = 13

color_Sn = '#e47a6b'  # 
color_Precision = '#ffa64d'  # 
color_Recall = '#94c4bb'  # 


dataset = ['Collins PIN', 'Gavin PIN', 'Krogan core PIN']

# 
methods = ['BOPS', 'CACO', 'CFinder', 'ClusterONE', 'Core&Peel', 'CPredictor', 'Copcc', 'DPClus', 'DPC-NADPIN', 'EWCA', 'GraphEntropy', 'IPCA', 'MCODE', 'SPICI', 'WCOACH', 'PCEGS']

Sn = [0.7068, 0.6, 0.6926, 0.5424, 0.4701, 0.4135, 0.5600, 0.5308, 0.6913, 0.5662, 0.5606, 0.3812, 0.1964, 0.4515, 0.522, 0.7825]
Precision = [0.5063, 0.7015, 0.2687, 0.4442, 0.3213, 0.6818, 0.6704, 0.6065, 0.6261, 0.6729, 0.6646, 0.72, 0.6538, 0.5664, 0.3955, 0.6426]
Recall = [0.7793, 0.6136, 0.7592, 0.6276, 0.4701, 0.4135, 0.6034, 0.5786, 0.6913, 0.6102, 0.5987, 0.3445, 0.1852, 0.4644, 0.5719, 0.8429]

# 2x2
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# 
bottom_values = np.zeros(len(methods))

# 
bars = []
for i, ax in enumerate(axes.flat[:-1]):  # 
    # 
    bar1 = ax.bar(methods, Sn, label='Sn',  width=0.7, color=color_Sn)
    bar2 = ax.bar(methods, Precision, bottom=Sn, label='Precision',width=0.7, color=color_Precision)
    bar3 = ax.bar(methods, Recall, bottom=np.array(Sn) + np.array(Precision), label='Recall', width=0.7, color=color_Recall)

    # 
    bars.append((bar1, bar2, bar3))

    # 
    ax.set_title(dataset[i])
    ax.set_ylabel('TS')
    # ax.set_xlabel('Methods')

    # 
    ax.set_xticklabels(methods, rotation=45, ha='right', fontsize=14)

# 
axes[1, 1].axis('off')  # 

# 
axes[1, 1].legend([bar[0] for bar in bars[0]] + [bar[1] for bar in bars[0]] + [bar[2] for bar in bars[0]], 
                   ['Sn', 'Precision', 'Recall'], 
                   loc='upper left', fontsize=14)

# 
plt.tight_layout()
plt.savefig('res.pdf')
# 
plt.show()
```
